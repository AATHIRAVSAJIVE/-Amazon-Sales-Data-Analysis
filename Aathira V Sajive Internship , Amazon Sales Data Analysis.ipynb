### <font color = blue> Feature Description: 

```python
# Number of numerical columns & Number of categorical columns in the DataFrame

numerical_columns = data.select_dtypes(include=['float', 'int']).columns
categorical_columns = data.select_dtypes(include=['object']).columns

print(f"Number of Numerical Columns: {len(numerical_columns)}")
print(f"Number of Categorical Columns: {len(categorical_columns)}")
```

1. Region  (categorical): Name of the geographical region where the sales transaction took place.
2. Country (categorical): Name of the specific country within the region where the sales transaction occurred.
3. Item Type (categorical): The type of item that was sold in the transaction. 
4. Sales Channel  (categorical): The channel through which the sale was made ( online / offline ).
5. Order Priority (categorical): The priority level of the order ( denoted by letters ('H' , 'M' , 'L' , 'C').
6. Order Date (temporal): The date when the order was placed.
7. Order ID (numerical): A unique identifier for each order.
8. Ship Date (temporal): The date when the ordered items were shipped. 
9. Units Sold (numerical): The number of units of the item sold in the transaction.  
10. Unit Price (numerical): The price of one unit of the item.
11. Unit Cost (numerical): The cost incurred to produce or acquire one unit of the item. 
12. Total Revenue (numerical): The total revenue generated from the sale, calculated as the product of Units Sold and Unit     Price.
13. Total Cost (numerical): The total cost incurred for the sale, calculated as the product of Units Sold and Unit Cost.
14. Total Profit (numerical): The total profit generated from the sale, calculated as the difference between Total Revenue     and Total Cost.

```python
# Check for Outliers:

# Select numeric columns
numeric_cols = data.select_dtypes(include=[np.number])

# Calculate the number of rows and columns for subplots
num_cols = len(numeric_cols.columns)
num_rows = int(np.ceil(num_cols / 2))

# Create a figure and axis for subplots
fig, axs = plt.subplots(nrows=num_rows, ncols=2, figsize=(8, 8))

# Flatten the axs array to simplify indexing
axs = axs.flatten()

# Iterate over each numeric column and create a box plot
for i, column in enumerate(numeric_cols.columns):
    sns.boxplot(y=numeric_cols[column], ax=axs[i])
    axs[i].set_title(column,color='brown', fontweight='bold')

# Hide empty subplots if there are an odd number of plots
if num_cols % 2 != 0:
    axs[-1].axis('off')

# Adjust the layout and spacing between subplots
fig.tight_layout()

# Show the plots
plt.show()
```

##### Detected outliers in the columns 'Total Revenue', 'Total Cost' & 'Total Profit' 

```python
# Select the columns of interest
columns_of_interest = ['Total Revenue', 'Total Cost', 'Total Profit']

# Iterate over each numeric column
for column in columns_of_interest:
    # Identify outliers
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Print the number of outliers
    outliers = (data[column] < lower_bound) | (data[column] > upper_bound)
    num_outliers = outliers.sum()
    print(f"The column '{column}' has {num_outliers} outliers.")

    # Print rows where outliers are present
    if num_outliers > 0:
        print(f"Rows with outliers in ")
        print(data.loc[outliers, [column]])
        print("\n")
```

```python
# Identify the top 5 products for the top 6 regions based on revenue

# Identify top 6 regions
top5_regions = data.groupby('Region')['Total Revenue'].sum().nlargest(6).index

# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16, 14))
fig.subplots_adjust(hspace=0.5)

# Create a bar graph for each region
for i, region in enumerate(top5_regions):
    row, col = divmod(i, 3)
    region_data = data[data['Region'] == region].groupby('Item Type')['Total Revenue'].sum().nlargest(5)
    region_data.plot(kind='bar', ax=axes[row, col], title=f'Top 5 Products in {region}')
    axes[row, col].set_title(f'Top 5 Products in {region}', fontweight='bold')  # Set fontweight here

# Show the plots
plt.show()
```

```python
#  Item sold the most in the year 2012

# Filter data for the year 2012
data_2012 = data[data['Order Date'].dt.year == 2012]

# Group by 'Item Type' and calculate the sum of 'Units Sold' for each item
item_sales_2012 = data_2012.groupby('Item Type')['Units Sold'].sum().reset_index()

# Sort the dataframe by 'Units Sold' in descending order to find the highest
item_sales_2012_sorted = item_sales_2012.sort_values(by='Units Sold', ascending=False)

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Create a bar plot with Seaborn
sns.barplot(x='Item Type', y='Units Sold', data=item_sales_2012_sorted, palette='tab20c')

# Set the title and labels
plt.title('Total Units Sold by Item Type in 2012')
plt.xlabel('Item Type')
plt.ylabel('Total Units Sold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

```python
# Create a pie chart showing the percentage of product categories by revenue

# Group by 'Item Type' and calculate the sum of 'Total Revenue'
revenue_by_category = data.groupby('Item Type')['Total Revenue'].sum().reset_index()

# Set up the matplotlib figure
plt.figure(figsize=(7, 7))

# Create a pie chart
plt.pie(revenue_by_category['Total Revenue'], labels=revenue_by_category['Item Type'], autopct='%1.1f%%', startangle=140)

# Set the title
plt.title('Percentage of Product Categories by Revenue',fontweight='bold')

# Display the pie chart
plt.show()
```

```python
#  Item with the highest order or sale

# Group by 'Item Type' and calculate the sum of 'Units Sold' for each item
item_sales = data.groupby('Item Type')['Units Sold'].sum().reset_index()

# Sort the dataframe by 'Units Sold' in descending order to find the highest
item_sales_sorted = item_sales.sort_values(by='Units Sold', ascending=False)

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Create a bar plot with Seaborn
sns.barplot(x='Item Type', y='Units Sold', data=item_sales_sorted, palette='tab20b')

# Set the title and labels
plt.title('Total Units Sold by Item Type',fontweight='bold')
plt.xlabel('Item Type')
plt.ylabel('Total Units Sold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()

# Filter data for the 'Cosmetics' item type
cosmetics_data = data[data['Item Type'] == 'Cosmetics']

# Extract month and year from 'Order Date'
cosmetics_data['MonthYear'] = cosmetics_data['Order Date'].dt.to_period('M')

# Group by 'MonthYear' and calculate the sum of 'Units Sold'
monthly_cosmetics_sales = cosmetics_data.groupby('MonthYear')['Units Sold'].sum().reset_index()

# Find the month and year with the highest sales
max_sales_month = monthly_cosmetics_sales.loc[monthly_cosmetics_sales['Units Sold'].idxmax()]

# Print the month and year with the highest sales
print(f"The month and year with the highest cosmetic sales: {max_sales_month['MonthYear']}")
```
